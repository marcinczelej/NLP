```
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=5000)
print("length {} | first 10 elements {}" .format(len(X_train), X_train[:10]))
print("length {} | first 10 elements {}" .format(len(y_train), y_train[:10]))
```

```
# getting word mapping, so they can be mapped back from number to word
words2idx = imdb.get_word_index()
idx2word = {i:word for word, i in words2idx.items()}
```

```
# padding input sequence to make the same size each
words_limit = 500

padded_train = sequence.pad_sequences(X_train, maxlen=words_limit, padding = "post")
padded_test = sequence.pad_sequences(X_test, maxlen=words_limit, padding="post")
```

```
def createModel(voacabulary_size, embedding_size):
  model = tf.keras.Sequential()
  model.add(layers.Embedding(input_dim=voacabulary_size, output_dim=embedding_size))
  model.add(layers.LSTM(128, dropout=0.3))
  model.add(layers.Dense(1, activation="sigmoid"))
  return model
```

```
model = createModel(5000,128)
model.summary()
```

```
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(padded_train, y_train, batch_size=32, epochs=5, verbose=2, validation_split=0.1)
```

```
model.evaluate(padded_test, y_test, verbose=1)
```
